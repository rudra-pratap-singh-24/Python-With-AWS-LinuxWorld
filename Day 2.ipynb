## IF- ELIF- ELSE

## CALCULATOR


```python
first_number= int(input("Enter the first number"))
second_number= int(input("Enter the second number"))
operator = input("Enter operation from *,-,+,/")

if operator == "+":
   print (first_number + second_number)
elif operator == "-":
   print (first_number - second_number)
elif operator == "*":
   print (first_number * second_number)
elif operator == "/":
   print (first_number / second_number)
else: 
   print ("Enter Valid option from *,-,+,/")
```

## Nested IF Condition

# ATM Code

```python
card_valid = True
correct_pin = 4321
account_balance = 100000

if card_valid == True:
    entered_pin = int(input("Enter your PIN: "))

    if entered_pin == correct_pin:
        withdraw_amount = int(input("Enter your withdrawal amount: "))

        if withdraw_amount <= account_balance:
            account_balance -= withdraw_amount
            print("Please collect your money.")
            print("Remaining Balance:", account_balance)
        else:
            print("Withdrawal amount is greater than account balance.")

    else:
        print("Enter Correct PIN")

else:
    print("Invalid Card")
```

# Loan approval System

```python
monthly_salary = int(input("Enter your monthly salary: "))
credit_score = int(input("Enter your credit score: "))
account_balance = int(input("Enter your account balance: "))

if monthly_salary >= 30000:
    if credit_score >= 700:
        if account_balance >= 200000:
            print("Loan Approved")
        else:
            print("Sorry! You have very less security deposit.")
    else:
        print("Sorry! Your credit score is low.")
else:
    print("Sorry! Loan not approved.")
```

# Password Attempts

```python
correct_password = "admin123"
attempt = 0

while attempt < 3:
    entered_password = input("Enter your password: ")

    if entered_password == correct_password:
        print("Login Success")
        break
    else:
        attempt += 1
        print("Wrong Password Entered. Try Again.")

if attempt == 3:
    print("Your account has been locked.")
```

# Counter Timer

```python
seconds = int(input("Enter the time in seconds: "))

while seconds >= 0:
    print(seconds)
    seconds = seconds - 1

print("Launch")
```
